# Decision Tree on the Iris data set

### Read in the data set and split into a training and testing set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split


df = pd.read_csv('./data/iris.csv')

features = df[['sepal_length', 'sepal_width',
               'petal_length', 'petal_width']].values

labels = df[['species']].values.reshape(1, -1)[0]

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.5)
```

### Create a scatter plot of the training and testing data for two features to get an idea of where the different clusters lie.

```python
%matplotlib inline
def plot_iris(X, y):
    labels = ['Iris-versicolor', 'Iris-setosa', 'Iris-virginica']
    colors = {'Iris-versicolor': 'red', 'Iris-setosa': 'blue', 'Iris-virginica': 'green'}

    for i in range(X.shape[0]):
        label = y[i]

        plt.scatter(X[i][0], X[i][1], color=colors[label])
        plt.xlabel('Sepal Length')
        plt.ylabel('Sepal Width')

plot_iris(X_train, y_train)
plt.title('A scatter plot of two features (Sepal Length and Sepal Width)\n for the Iris training data set.')
plt.show()

plot_iris(X_test, y_test)
plt.title('A scatter plot of two features (Sepal Length and Sepal Width)\n for the Iris testing data set.')
plt.show()
```

### Train a decision tree model on the training set

```python
from decision_tree import DecisionTree

model = DecisionTree()
model.fit(X_train, y_train)

train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)
```

### Evaluate the accuracy on the both the training and testing sets.

```python
train_accuracy = model.accuracy(y_train, train_predictions)
test_accuracy = model.accuracy(y_test, test_predictions)

print("Train Accuracy: %.2f" % train_accuracy)
print("Test Accuracy: %.2f" % test_accuracy)
```

### Create a scatter plot of the testing data for both the TRUE and PREDICTED labels

```python
plot_iris(X_test, y_test)
plt.title('A scatter plot of two features (Sepal Length and Sepal Width)\n for testing data set (true labels)')
plt.show()

plot_iris(X_test, test_predictions)
plt.title('A scatter plot of two features (Sepal Length and Sepal Width)\n for testing data set (predictions)')
plt.show()
```
